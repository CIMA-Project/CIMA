```python
import pandas as pd
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
plt.rcParams["font.family"]="sans-serif"
plt.rcParams["pdf.fonttype"]= 42
```

```python
genotype = pd.read_parquet('/CIMA/Data/413_sample_genotype.parquet')
genotype = genotype.T
```

```python
def generate_plot_df(file_path,probe,snp): 
    phenotype = pd.read_csv(file_path,index_col=0)
    phenotype = phenotype[probe]
    genotype_selection = genotype.loc[phenotype.index,snp]
    plot_df = pd.DataFrame({'variation':genotype_selection,'phenotype':phenotype})
    return(plot_df)
```

```python
list_detected_IKZF4 = ['CD4_Tem_CCR7neg', 'CD4_Tfh-like_CXCR5', 'CD4_Th1-like_GZMK', 'CD4_Th17-like_RORC','CD4_Tn_CCR7', 'CD4_Treg_FCRL3', 'CD4_Treg_FOXP3', 'CD8_CTL_GZMB','CD8_Tem_CCR7neg', 'CD8_Tn_CCR7', 'cMono_CD14', 'ncMono_FCGR3A']
```

```python
for celltype in list_detected_IKZF4:
    plot_df = generate_plot_df(f'/CIMA/Data/eQTL/normal_dis/{celltype}.csv','IKZF4','chr12_56050848')
    plt.figure(figsize=(1.5, 2))
    sns.set_theme(style="ticks")
    sns.violinplot(x='variation', y='phenotype', data=plot_df,color = '#95AFBA', inner='box')
    sns.despine()
    plt.xlabel('CC-CT-TT')
    plt.ylabel('IKZF4')
    plt.title(f'{celltype}')
    plt.grid(False)
    # xy
    plt.savefig(f'/CIMA/Result/plot/box_vio_plot/20250312_IKZF4_in_{celltype}.pdf', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
merge_df_beta_P = pd.DataFrame()

for celltype in list_detected_BLK:
    a1 = pd.read_parquet(f'/CIMA/Result/eQTL_tensorqtl_mergenominal/{celltype}.parquet')

    a1['pair'] = a1['phenotype_id']+'_'+a1['variant_id']

    a1 = a1[a1['pair'] == 'BLK_chr8_11480861']

    a1['celltype'] = celltype
    merge_df_beta_P = pd.concat([merge_df_beta_P,a1])

merge_df_beta_P.to_csv('/CIMA/Result/plot/box_vio_plot/BLK/20250322_BLK_beta_p.csv')
```
